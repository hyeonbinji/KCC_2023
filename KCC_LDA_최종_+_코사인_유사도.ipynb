```python
def text_preprocessor(s):
    import re
    
    ## (1) [], (), {}, <> 괄호와 괄호 안 문자 제거하기
    pattern = r'\([^)]*\)'  # ()
    s = re.sub(pattern=pattern, repl='', string=str(s))
    
    pattern = r'\[[^)]*\]'  # []
    
    s = re.sub(pattern=pattern, repl='', string=str(s))
    
    pattern = r'\<[^)]*\>'  # <>
    s = re.sub(pattern=pattern, repl='', string=str(s))
    
    pattern = r'\{[^)]*\}'  # {}
    s = re.sub(pattern=pattern, repl='', string=str(s))
    
    
    
    ## (2) '.'제거하기
    s = s.replace('.', ' ')
    s = s.replace(',', ' ')

    
    ## (3) 특수문자 제거
    pattern = r'[^a-zA-Z가-힣]'
    s = re.sub(pattern=pattern, repl=' ', string=str(s))
    #일단 숫자도 다 제거

            
    return s


#clo: 띄어쓰기 없애는 컬럼
def clo(x):
    x = ' '.join(x.split())
    
    return x

#prepro: 한글자 제외하고 맞춤법 검사
def prepro(x):
  
    spelled_sent = spell_checker.check(x)
    hanspell_sent = spelled_sent.checked
    
    s_split = hanspell_sent.split()
    
    #글자 1개만 있으면 제외하기
    s_list = []
    for word in s_split:
        if len(word) !=1:
            s_list.append(word)
            
    return s_list
    return(hanspell_sent)

#토큰화된거 다시 합치는 함수
def last(x):
  return ' '.join(x)
```

```python
def make_topictable_per_doc(ldamodel, corpus):
    topic_table = pd.DataFrame()

    # 몇 번째 문서인지를 의미하는 문서 번호와 해당 문서의 토픽 비중을 한 줄씩 꺼내온다.
    for i, topic_list in enumerate(ldamodel[corpus]):
        doc = topic_list[0] if ldamodel.per_word_topics else topic_list            
        doc = sorted(doc, key=lambda x: (x[1]), reverse=True)
        # 각 문서에 대해서 비중이 높은 토픽순으로 토픽을 정렬한다.
        # EX) 정렬 전 0번 문서 : (2번 토픽, 48.5%), (8번 토픽, 25%), (10번 토픽, 5%), (12번 토픽, 21.5%), 
        # Ex) 정렬 후 0번 문서 : (2번 토픽, 48.5%), (8번 토픽, 25%), (12번 토픽, 21.5%), (10번 토픽, 5%)
        # 48 > 25 > 21 > 5 순으로 정렬이 된 것.

        # 모든 문서에 대해서 각각 아래를 수행
        for j, (topic_num, prop_topic) in enumerate(doc): #  몇 번 토픽인지와 비중을 나눠서 저장한다.
            if j == 0:  # 정렬을 한 상태이므로 가장 앞에 있는 것이 가장 비중이 높은 토픽
                topic_table = topic_table.append(pd.Series([int(topic_num), round(prop_topic,4), topic_list]), ignore_index=True)
                # 가장 비중이 높은 토픽과, 가장 비중이 높은 토픽의 비중과, 전체 토픽의 비중을 저장한다.
            else:
                break
    return(topic_table)
```

```python
def LDA_reason(data):
  res = data['고장 원인']
  df_all_res = pd.DataFrame(res)

  #간단한 전처리
  df_all_res['preprocessed'] = df_all_res['고장 원인'].apply(lambda s: text_preprocessor(s))
  #띄어쓰기 없앤다.
  df_all_res['preprocessed_clo'] = df_all_res['preprocessed'].apply(lambda s: clo(s))

#다시 spacing 으로 맞춤법 및 띄어쓰기 검사

  df_all_res.reset_index(inplace = True, drop = True)


  from pykospacing import Spacing
  for i, j in enumerate(df_all_res['preprocessed_clo']):
      spacing = Spacing()
      kospacing_sent = spacing(df_all_res['preprocessed_clo'][i])
   
      df_all_res['preprocessed_clo'][i] = kospacing_sent



  okt = Okt()

  ls_prepro = list(df_all_res['preprocessed_clo'])
  noun_ls_res = []
  for i in range(len(ls_prepro)):
      noun_ls_res.append(okt.nouns(ls_prepro[i]))


  tokenized = []

  for doc in noun_ls_res:
      a_ls = []
      for word in doc:
          if len(word) == 1:
              continue
          else:
              a_ls.append(word)
      tokenized.append(list(a_ls))

#토픽모델링 딕셔너리, 코퍼스 만들기
  global dictionary
  global corpus
  dictionary = corpora.Dictionary(tokenized)
  corpus = [dictionary.doc2bow(text) for text in tokenized]


#### coherence, perplexity score 계산 후 시각화
  min_topics = 1
  max_topics = 15
  step_size = 1


  topics_range = range(min_topics, max_topics+1, step_size)
  coherence_scores = []
  perplexity_scores = []


  for num_topics in topics_range:
    # Train LDA model
      lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, random_state = 100,
                         passes=20)
    
    # coherence score
      coherence_model_lda = CoherenceModel(model=lda_model, texts=tokenized,
                                         dictionary=dictionary, coherence='c_v')
      coherence_score = coherence_model_lda.get_coherence()
      coherence_scores.append(coherence_score)


    # perplexity score
      perplexity_score = lda_model.log_perplexity(corpus)
      perplexity_scores.append(-perplexity_score) #log를 취한 값은 높을수록 좋다. 그렇기에 -를 붙여 낮을수록 좋은 perplexity 시각화.

# Plot results
  fig, ax1 = plt.subplots()

  color = 'tab:blue'
  ax1.set_xlabel('Number of Topics')
  ax1.set_ylabel('Coherence Score', color=color)
  ax1.plot(topics_range, coherence_scores, color=color)
  ax1.tick_params(axis='y', labelcolor=color)

  ax2 = ax1.twinx()  

  color = 'tab:red'
  ax2.set_ylabel('Perplexity Score', color=color)  
  ax2.plot(topics_range, perplexity_scores, color=color)
  ax2.tick_params(axis='y', labelcolor=color)

  fig.tight_layout() 
  plt.show()
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_accident(df, input_text, top_n=10):
    """
    df: 데이터프레임
    input_text: 비교하고자 하는 텍스트
    top_n: 추출하고자 하는 결과의 개수 (기본값: 5)
    """
    df = df.fillna('')

    # TF-IDF 벡터화
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(df['고장 원인'])

    # 입력 텍스트 벡터화
    input_tfidf = tfidf_vectorizer.transform([input_text])

    # 코사인 유사도 계산
    cosine_similarities = cosine_similarity(input_tfidf, tfidf_matrix).flatten()

    # similarities 리스트에서 내림차순으로 정렬하여 상위 5개의 인덱스 추출
    similar_indices = np.argsort(cosine_similarities)[::-1][1:10]

    # 결과 데이터프레임의 인덱스를 추출한 인덱스로 설정
    result_df = df.iloc[similar_indices, :]
    result_df = result_df.reset_index(drop=True)  # 인덱스 재설정

    result_df['cos_similarity'] = [cosine_similarities[idx] for idx in similar_indices]

    return result_df
```
