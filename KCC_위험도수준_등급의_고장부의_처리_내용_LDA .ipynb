```python
def text_preprocessor(s):
    import re
    
    ## (1) [], (), {}, <> 괄호와 괄호 안 문자 제거하기
    pattern = r'\([^)]*\)'  # ()
    s = re.sub(pattern=pattern, repl='', string=str(s))
    
    pattern = r'\[[^)]*\]'  # []
    s = re.sub(pattern=pattern, repl='', string=str(s))
    
    pattern = r'\<[^)]*\>'  # <>
    s = re.sub(pattern=pattern, repl='', string=str(s))
    
    pattern = r'\{[^)]*\}'  # {}
    s = re.sub(pattern=pattern, repl='', string=str(s))
    
    
    
    ## (2) '.'제거하기
    s = s.replace('.', ' ')
    s = s.replace(',', ' ')

    
    ## (3) 특수문자 제거
    pattern = r'[^a-zA-Z가-힣]'
    s = re.sub(pattern=pattern, repl=' ', string=str(s))
    #일단 숫자도 다 제거

            
    return s


#clo: 띄어쓰기 없애는 컬럼
def clo(x):
    x = ' '.join(x.split())
    
    return x

#prepro: 한글자 제외하고 맞춤법 검사
def prepro(x):
  
    spelled_sent = spell_checker.check(x)
    hanspell_sent = spelled_sent.checked
    
    s_split = hanspell_sent.split()
    
    #글자 1개만 있으면 제외하기
    s_list = []
    for word in s_split:
        if len(word) !=1:
            s_list.append(word)
            
    return s_list
    return(hanspell_sent)

#토큰화된거 다시 합치는 함수
def last(x):
  return ' '.join(x)
```

##### 고 - 고장부의

```python
res = final_high_txt['고장 부의']
df_all_res = pd.DataFrame(res)

#간단한 전처리
df_all_res['preprocessed'] = df_all_res['고장 부의'].apply(lambda s: text_preprocessor(s))
#띄어쓰기 없앤다.
df_all_res['preprocessed_clo'] = df_all_res['preprocessed'].apply(lambda s: clo(s))

#다시 spacing 으로 맞춤법 및 띄어쓰기 검사

df_all_res.reset_index(inplace = True, drop = True)


from pykospacing import Spacing
for i, j in enumerate(df_all_res['preprocessed_clo']):
    spacing = Spacing()
    kospacing_sent = spacing(df_all_res['preprocessed_clo'][i])
   
    df_all_res['preprocessed_clo'][i] = kospacing_sent



okt = Okt()

ls_prepro = list(df_all_res['preprocessed_clo'])
noun_ls_res = []
for i in range(len(ls_prepro)):
    noun_ls_res.append(okt.nouns(ls_prepro[i]))


tokenized = []

for doc in noun_ls_res:
    a_ls = []
    for word in doc:
        if len(word) == 1:
            continue
        else:
            a_ls.append(word)
    tokenized.append(list(a_ls))

#토픽모델링 딕셔너리, 코퍼스 만들기
dictionary = corpora.Dictionary(tokenized)
corpus = [dictionary.doc2bow(text) for text in tokenized]



max_topics = 20

# 토픽 수에 따른 연관성 점수 리스트
coherence_scores = []

for num_topics in range(1, max_topics+1):
    # LDA 모델 학습
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                                id2word=dictionary,
                                                num_topics=num_topics,
                                                random_state=0,
                                                chunksize=100,
                                                passes=10)

    # 연관성 점수 계산
    coherence_model_lda = CoherenceModel(model=lda_model, texts=tokenized, dictionary=dictionary, coherence='c_v')
    coherence_score = coherence_model_lda.get_coherence()
    coherence_scores.append(coherence_score)


# 연관성 점수 시각화
import matplotlib.pyplot as plt
plt.plot(range(1, max_topics+1), coherence_scores)
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score")
plt.show()

x = range(1, max_topics+1, 1)
best_num_topics = x[coherence_scores.index(max(coherence_scores))]
```

```python
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                            id2word=dictionary,
                                            num_topics=2,
                                            random_state=42,
                                            update_every=1,
                                            chunksize=100,
                                            passes=10,
                                            alpha='auto',
                                            per_word_topics=True)

pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)
```
